## Task 2

## Univariate Analysis:

Univariate analysis is the simplest form of analysis, and it can be done on both numerical and categorical features. Here's how to get started with univariate analysis:

For numerical features, you can create summary statistics, histograms, box plots, and density plots to understand the distribution and patterns

For categorical features, you can create bar plots to visualize the distribution of categories.

### Univariate analysis for numerical features

```python
# Summary statistics for a numerical variable
summary = dataset[['LotArea','OverallQual','YrSold','SalePrice']].describe()
print(summary)
```

```python
numerical_columns = ['LotArea', 'OverallQual', 'YrSold', 'SalePrice']

# Create separate histograms for each numerical variable
for column in numerical_columns:
    plt.figure(figsize=(4, 3))
    sns.histplot(dataset[column], kde=True)
    plt.title(f"Histogram of {column}")
    plt.show()
```

### According to the histograms;
1. Lot area of most of the houses fall within the 0 to 50000 square feet bucket
2. Overall quality of most of the houses is 5 i.e., average
3. There were lesser sale in the year of 2010 as compare to years 2006 to 2009
4. Sale price of most of the houses fall near 200,000.00

```python
# Box plot for a numerical variable
plt.figure(figsize=(4, 3))
sns.boxplot(x=dataset['SalePrice'])
plt.title("Box Plot of Sale Price")
plt.show()
```

### According to the Box Plot of sale price, there are so many outliers in sale price

```python
from scipy.stats import gaussian_kde

# Create a density plot
density = gaussian_kde(dataset['YrSold'])
x_values = range(int(min(dataset['YrSold'])), int(max(dataset['YrSold'])) + 1)
y_values = density(x_values)

plt.plot(x_values, y_values)
plt.fill_between(x_values, y_values, alpha=0.2)
plt.title("Density Plot")
plt.xlabel("Year Sold")
plt.ylabel("Density")
plt.show()
```

### According to the Density Plot of year sold, in 2009, highest number of units sold

### Univariate analysis for categorical features

```python
# Count the occurrences of each category
counts = {}
for item in dataset['MSZoning']:
    counts[item] = counts.get(item, 0) + 1

# Extract category names and counts
categories = list(counts.keys())
counts = list(counts.values())

# Create a bar plot
plt.bar(categories, counts)
plt.title("Bar Plot for MSZoning")
plt.xlabel("MSZoning")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.show()
```

### According to the Bar Plot of MSZoning, RL i.e., Residential Low-Density has the highest number of units

```python
# Bar plot for a categorical variable
plt.figure(figsize=(4, 3))
sns.countplot(data=dataset, x='MSSubClass')
plt.title("Bar Plot of MSSubClass")
plt.xticks(rotation=45)
plt.show()
```

### According to the Bar Plot of MSSubClass, SC20 i.e., 1-STORY 1946 & NEWER ALL STYLES has the highest number of units

```python
# Scatter plot for two numerical variables
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dataset, x='YrSold', y='SalePrice')
plt.title("Scatter Plot of YrSold vs. SalePrice")
plt.show()
```

### According to the Scatter plot for Sale Price and Year Sold, houses with highest price were sold in the year of 2007

## Task 3

## Multivariate Analysis:

Multivariate analysis involves exploring relationships between multiple variables. You can use various plots and statistical tests to analyze these relationships.

Here are some examples:

For numeric vs. numeric relationships, you can create scatter plots, correlation matrices, and regression analysis.

For numeric vs. categorical relationships, you can create box plots, violin plots, and t-tests.

For categorical vs. categorical relationships, you can create contingency tables, chi-squared tests, and mosaic plots.

### Multivariate analysis numeric vs. numeric relationships

```python
plt.figure(figsize=(2, 2))
sns.heatmap(dataset[['YrSold','SalePrice']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### According to the Correlation Matrix of Year Sold and Sale Price, there is a no relationship between the Year Sold and Sale Price

```python
import statsmodels.api as sm

# Sample data
x = dataset['YrSold']
y = dataset['SalePrice']

# Add a constant to the independent variable for the intercept term
x = sm.add_constant(x)

# Fit the regression model
model = sm.OLS(y, x).fit()

# Print the summary statistics
print(model.summary())
```

### Multivariate analysis numeric vs. categorical

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

# Create a box plot
sns.boxplot(x="MSSubClass", y="YrSold", data=dataset)
plt.title("Box Plot")

plt.show()
```

### According to the Box Plot of Year Sold and MSSubClass, most type of the house were sold between 2007 and 2009 except SC190, SC160, SC75, SC180 and SC40. Sale of SC190 and SC75 happened between 2006 and 2008

```python
# Create a violine plot
sns.violinplot(x=dataset["YrSold"], y=dataset["SalePrice"], data=dataset, hue=dataset["MSZoning"], split=False)
plt.title("Side-by-Side Violin Plot of Sale Price by Year Sold with MS Zoning")
plt.show()
```

### According to the Violine Plot of Year Sold and SalePrice, there were no sale of MSZoning C (all) i.e., Commercial in the year of 2007

```python
import scipy.stats as stats

# data for two independent groups
group1 = dataset['YearBuilt']
group2 = dataset['YrSold']

# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(group1, group2)

# Print the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Determine if the difference is statistically significant (e.g., using a significance level of 0.05)
if p_value < 0.05:
    print("The means are significantly different.")
else:
    print("There is no significant difference between the means.")
```

## Analysis of Data with Plots:

You can use various types of plots to analyze your data, including:

Box plots: Visualize the distribution of data and identify outliers.

Scatter plots: Explore relationships between two numerical variables.

Violin plots: Show the distribution of data across different categories.

Bubble plots: Visualize three-dimensional data with the size and color of points.

Histograms: Display the frequency distribution of a single numerical variable.

Bar plots: Visualize the distribution of categories in a categorical variable.

Kernel Density Estimation (KDE): Smoothed representation of data distribution.

```python
# Create a Box plot for MS Zoning
plt.figure(figsize=(8, 6))
sns.boxplot(x=dataset['MSZoning'], y=dataset['SalePrice'])
plt.title("Box Plot of SalePrice by MSZoning")
plt.show()
```

### Accoording to the Box Plot for MS Zoning and Sale Price, there are no outliers in sale price of C(all) i.e., commercial and RH i.e., Residential High-Density. There are more outliers in RL i.e., Residential Low-Density as compare to RM i.e., Residential Medium Density. There are very few outliers in FV i.e., Floating Village Residential

```python
# Scatter plot for sale year built and year sold
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dataset, x='YearBuilt', y='YrSold')
plt.title("Scatter Plot of YearBuilt vs. YrSold")
plt.show()
```

### According to the Scatter plot for sale year built and year sold, most of houses were buit after 1990.

```python
# Create a violin plot
sns.violinplot(x="MSZoning", y="SalePrice", data=dataset)
plt.show()
```

### According to the Violin Plot, Sale price of MSZoning RL i.e., Residential Low-Density is highest and sale price of C(all) i.e., Commercial is lowest

```python
# Create a bubble plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=dataset['MSSubClass'], y=dataset['MSZoning'], hue=dataset['OverallQual'], edgecolors='black', linewidth=1.5)

# Add labels and title
plt.title('Bubble Plot MSSubClass and MSZoning')
plt.xlabel('MSSubClass')
plt.ylabel('MSZoning')

# Show the plot
plt.show()
```

### According to the Bubble Plot for MS Zoning and MS SubClass, overall quality of MSSubClass SC60 i.e., 2-STORY 1946 & NEWER in MSZoning FV i.e., Floating Village Residential is 10 i.e., very excellent. Overall quality of SC20 i.e., 1-STORY 1946 & NEWER ALL STYLES , SC120 i.e., 1-STORY PUD (Planned Unit Development) - 1946 & NEWER and SC160 i.e., 2-STORY PUD - 1946 & NEWER in FV is 7 i.e., good.

```python
# Create a Histograms
plt.hist(dataset['SaleType'], bins=30, edgecolor='black')

# Add labels and title
plt.title('Histogram for Sale Type')
plt.xlabel('SaleType')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
